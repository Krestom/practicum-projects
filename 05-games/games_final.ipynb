# Рейтинг ESRB.
    
**Для удобства перечислю описание категорий рейтинга, который встречается в данных:**
    
    
  > **EC** - Для детей младшего возраста. Подходит для детей от 3 лет, неприемлемое содержание отсутствует. Примеры: Дора Исследовательница, Сказки Дракона.  
    
  > **E** - Подходит для всех возрастных категорий. Игра не должна содержать звуков или изображений, которые могут напугать маленьких детей. Не должна использоваться ненормативная лексика. Примеры: Просто таннцуй, FIFA.
    
  > **E10+** - Подходит для лиц старше 10 лет. Игра может содержать умеренные формы насилия, а некоторые сцены могут пугать детей. Примеры: Minecraft Dungeons, Растения против зомби.
    
  > **T** - Подходит для лиц старше 13 лет. Игра может содержать более реалистичные и натуралистические сцены насилия. Примеры: Fortnite, Sims 4. 
    
  > **M** - Подходит для лиц старше 17 лет. Этот рейтинг используется, когда насилие становится реалистичным и похожим на реальную жизнь. Также может присутствовать ненормативная лексика, употребление табака, алкоголя или запрещенных наркотиков. Примеры: Ark: Survival Evolved, Destiny 2.  
    
  > **AO** - Подходит для взрослых старше 18 лет. Классификация «для взрослых» используется, когда в игре присутствует крайний уровень насилия и немотивированного убийства. Также может присутствовать романтизация наркотиков, азартных игр и сексуальной активности. Примеры: Grand Theft Auto V, Fallout 4.
    
  > **RP** - Категория ещё не присвоена. Играм с категорией RP еще не присвоен окончательный рейтинг ESRB.  
    
  > **K-A** - Игры, содержащие контент, подходящий для широкой аудитории. Этот рейтинг использовался до 1998 года, когда он был переименован в E.

**Теперь посмотрим, есть ли такая же взаимосвязь на других платформах.**

Создадим функцию, которая принимает в себя название платформы и возвращает графики корреляции и коеффициент корреляции.

```python
# Обьявление функции
def correlation(name_of_platform):
    platform = df_actual[df_actual['platform'] == name_of_platform]
    
    # создание фигуры и общего заголовка
    plt.figure(figsize=(15, 7))
    plt.suptitle(str.upper(name_of_platform), fontsize=30, color='blue')
    
    # Построение первого графика
    plt.subplot(1, 2, 1)
    sns.scatterplot(data=platform, x='total_sales', y='user_score')
    plt.title('Диаграмма рассеяния для оценок пользователей', fontsize=20)
    plt.xlabel('Продажи', fontsize=14)
    plt.ylabel('Оценки пользователей', fontsize=14)
    
    # Построение второго графика
    plt.subplot(1, 2, 2)
    sns.scatterplot(data=platform, x='total_sales', y='critic_score', color='coral')
    plt.title('Диаграмма рассеяния для оценок критиков', fontsize=20)
    plt.xlabel('Продажи', fontsize=14)
    plt.ylabel('Оценки критиков', fontsize=14)

    plt.tight_layout()
    plt.show()
    
    # Считаем корреляции и сохраняем в переменные
    user_corr = platform['total_sales'].corr(platform['user_score'])
    critic_corr = platform['total_sales'].corr(platform['critic_score'])
    
    # Выводим переменные с помощью f-строки
    if user_corr == 0:
        print(f'Связь между продажами и оценками пользователей отсутствует: {user_corr}')
    elif 0.001 <= user_corr <= 0.29:
        print(f'Связь между продажами и оценками пользователей слабая положительная: {user_corr}')
    elif 0.3 <= user_corr <= 0.69:
        print(f'Связь между продажами и оценками пользователей умеренная положительная: {user_corr}')
    elif 0.7 <= user_corr <= 1:
        print(f'Связь между продажами и оценками пользователей сильная положительная: {user_corr}')
    elif -0.29 <= user_corr <= -0.01:
        print(f'Связь между продажами и оценками пользователей слабая отрицательная: {user_corr}')
    elif -0.69 <= user_corr <= -0.3:
        print(f'Связь между продажами и оценками пользователей умеренная отрицательная: {user_corr}')
    elif -1 <= user_corr <= -0.7:
        print(f'Связь между продажами и оценками пользователей сильная отрицательная: {user_corr}')
    else:
        print(f'Связь между продажами и оценками пользователей слишком мала: {user_corr}')
     
    if critic_corr == 0:
        print(f'Связь между продажами и оценками критиков отсутствует: {critic_corr}')
    elif 0.001 <= critic_corr <= 0.29:
        print(f'Связь между продажами и оценками критиков слабая положительная: {critic_corr}')
    elif 0.3 <= critic_corr <= 0.69:
        print(f'Связь между продажами и оценками критиков умеренная положительная: {critic_corr}')
    elif 0.7 <= critic_corr <= 1:
        print(f'Связь между продажами и оценками критиков сильная положительная: {critic_corr}')
    elif -0.29 <= critic_corr <= -0.01:
        print(f'Связь между продажами и оценками критиков слабая отрицательная: {critic_corr}')
    elif -0.69 <= critic_corr <= -0.3:
        print(f'Связь между продажами и оценками критиков умеренная отрицательная: {critic_corr}')
    elif -1 <= critic_corr <= -0.7:
        print(f'Связь между продажами и оценками критиков сильная отрицательная: {critic_corr}')
```

##  Портрет пользователя каждого региона.

**Создадим функцию, создающую портрет пользователя по заданному региону.**

```python
# Обьявление функции с двумя параметрами, интересующий регион и его название
def portrait(sales_per_region, name_of_region):
    # Создадим сводную таблицу по платформе и ее продажам
    region_pivot = (df_actual
                    .pivot_table(index='platform', values=sales_per_region, aggfunc='sum')
                    .reset_index()
                    .sort_values(by=sales_per_region, ascending=False)
                    .head()
                   )
    
    # Сохраним в переменные название платформы и продажи
    platform = region_pivot['platform'].str.upper()
    region_sales = region_pivot[sales_per_region]
    
    # Создадим цвета для долей
    colors = plt.cm.tab20.colors
    
    # Построим график
    plt.figure(figsize=(10, 6))
    plt.pie(region_sales, labels=platform, autopct='%1.1f%%', startangle=90, colors=colors, shadow=True)
    plt.axis('equal')
    plt.title(f'Популярные платформы в {name_of_region}', fontsize=20, pad=20)
    plt.show()
```

## Проверка гипотез.

**Гипотеза №1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

Так как у нас тест со знаком !=, то он будет двусторонним.

Сформулируем нулевую и альтернативную гипотезы и проведем статистический тест.

*H<sub>0</sub>*: Средние пользовательские рейтинги платформ Xbox One и PC равны;

*H<sub>1</sub>*: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
print(df_actual[df_actual['platform'] == 'xone']['user_score'].isna().sum())
df_actual[df_actual['platform'] == 'pc']['user_score'].isna().sum()
```

В данных есть пропуски, удалим эти значения, чтобы провести статистический тест

```python
# Сохраним в переменные данные о рейтинге по платформе
xone_rating = df_actual[df_actual['platform'] == 'xone']['user_score'].dropna()
pc_rating = df_actual[df_actual['platform'] == 'pc']['user_score'].dropna()

# Установим уровень статистической значимости
alpha = 0.01

# Проведем статистический тест
results = st.ttest_ind(xone_rating, pc_rating, equal_var=False) # Без e_v=False pvalue  так же близко к нулю

# Выведем p-value
print(f"p-значение: {results.pvalue}")

# Проверим гипотезы
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC различаются.')
else:
    print('Не можем отвергнуть нулевую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC равны.')
```

**Гипотеза №2: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.**

Так как у нас тест со знаком !=, то он будет двусторонним.

Сформулируем нулевую и альтернативную гипотезы и проведем статистический тест.

*H<sub>0</sub>*: Средние пользовательские рейтинги жанров Action и Sports равны;

*H<sub>1</sub>*: Средние пользовательские рейтинги жанров Action и Sports не равны.

```python
# Сохраним в переменные данные по пользовательскому рейтингу и жанру
action_rating = df_actual[df_actual['genre'] == 'action']['user_score'].dropna()
sports_rating = df_actual[df_actual['genre'] == 'sports']['user_score'].dropna()

# Установим уровень статистической значимости
alpha = 0.01

# Проведем тест
results = st.ttest_ind(action_rating, sports_rating, equal_var=False)

# Выведем p-value
print(f'p-значение: {results.pvalue}')

# Проверим гипотезы
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports различаются.')
else:
    print('Не можем отвергнуть нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports равны.')
```
